#### Concepto Clave: Análisis de Correlación

El **Análisis de Correlación** es una herramienta esencial del Análisis Exploratorio de Datos (AED)  que me permite medir la relación que existe entre dos variables numéricas. Esto es crucial porque los modelos de Inteligencia Artificial se ven perjudicados si reciben información duplicada o redundante.

Mi objetivo es detectar si dos variables miden el mismo concepto de la vida real. Si lo hacen, una de las dos es candidata a ser eliminada para evitar que el modelo asuma que esa información es el doble de importante.

##### Tipos de Correlación

El resultado de la correlación se mide con un valor que oscila entre **-1** y **+1**:

*   **Correlación Positiva Fuerte ($\approx +1$):**
    *   Si una variable **aumenta**, la otra variable también **aumenta** .
    *   En un gráfico de dispersión (*Scatter Plot*), los puntos se alinearían en una línea recta ascendente.
    *   **Ejemplo:** A medida que la longitud de la aleta de un pingüino es mayor, su masa corporal (peso) también es mayor.

*   **Correlación Negativa Fuerte ($\approx -1$):**
    *   Si una variable **aumenta**, la otra variable **disminuye**.
    *   En un *Scatter Plot*, los puntos se alinearían en una línea recta descendente.

*   **Correlación Baja o Nula ($\approx 0$):**
    *   No hay una relación clara entre las variables.
    *   Los puntos se dispersan sin seguir un patrón.

#### Criterio de Eliminación

Cuando encuentro una correlación muy fuerte (por encima de **+0.8** o por debajo de **-0.8**), borro una de las columnas en la UD2.

*   **Decisión:** Se borra la variable que sea más **cara** o más **difícil de conseguir** o la que tenga menos **explicabilidad**.

### Análisis de Correlación y Mapa de Calor

El **Análisis de Correlación** mide la relación lineal entre dos variables numéricas, oscilando entre -1 y +1. Es fundamental en el **AED** (Análisis Exploratorio de Datos) para identificar columnas redundantes.

###  Código del Mapa de Calor (*Heatmap*)

Para visualizar la correlación, utilizaré la función `.corr()` de **Pandas** y la representaré

```python
# 1. Calculamos la matriz de correlación solo para variables numéricas.
# La función .corr() calcula el coeficiente de correlación de Pearson.
matriz_correlacion = df_titanic.corr(numeric_only=True)

# 2. Creamos la figura (lienzo) para el gráfico.
plt.figure(figsize=(10, 8))

# 3. Generamos el Mapa de Calor (Heatmap)
# 'annot=True' muestra los valores del coeficiente en cada celda (muy importante).
# 'cmap="coolwarm"' usa una paleta de colores divergente (rojo/azul) que facilita
# la identificación de correlaciones positivas y negativas.
sns.heatmap(
    matriz_correlacion,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=.5, # Espacio entre celdas para mayor claridad
    cbar_kws={'label': 'Coeficiente de Correlación'}
)

plt.title('Mapa de Calor de la Correlación entre Variables Numéricas')
plt.show()
```

--------------------------------------------------------------------------------

El mapa de calor me muestra el coeficiente de correlación de Pearson, que oscila entre -1 y +1. Este análisis es fundamental en el AED para detectar variables redundantes o altamente relacionadas.

Mi criterio es buscar correlaciones **muy fuertes** ($\approx \pm 1$) para justificar la eliminación de columnas.

###### Correlaciones entre variables independientes:

## Estudio y Detección de Valores Atípicos (Outliers)

El último paso del Análisis Exploratorio de Datos (AED) es identificar los **valores atípicos**. Un valor atípico es un dato que se desvía significativamente de la mayoría de las observaciones.

Mi objetivo es detectar estos valores para informar sobre ellos. El tratamiento de estos valores (eliminarlos o manipularlos) se llevará a cabo en la **UD2: Preprocesamiento de la información**.




### Detección Analítica en la Tarifa (`Fare`)

Utilizaré el método del **Rango Intercuartílico (IQR)** para detectar matemáticamente los *outliers*. Este es el método más común para definir los límites de los bigotes en el diagrama de caja (*boxplot*).

**Fórmula de Límite Superior:** $Q3 + 1.5 \times IQR$

```python
# 1. Definimos la columna a analizar: Fare (Tarifa), ya que en el boxplot se vieron atípicos.
columna_analizar = df_titanic['fare']

# 2. Calculamos los cuartiles Q1 (25%) y Q3 (75%).
Q1 = columna_analizar.quantile(0.25)
Q3 = columna_analizar.quantile(0.75)

# 3. Calculamos el Rango Intercuartílico (IQR)
IQR = Q3 - Q1

# 4. Definimos los límites para considerar un dato como atípico (Outlier).
# Usaremos el límite superior, ya que el boxplot mostró atípicos en la parte alta.
limite_superior = Q3 + 1.5 * IQR
# limite_inferior = Q1 - 1.5 * IQR

print(f"Cuartil 1 (Q1): {Q1:.2f}")
print(f"Cuartil 3 (Q3): {Q3:.2f}")
print(f"Rango Intercuartílico (IQR): {IQR:.2f}")
print(f"Límite Superior para Outliers (Q3 + 1.5 * IQR): {limite_superior:.2f}")
```

## Conteo y Conclusión
Ahora utilizo una máscara booleana (una técnica muy útil en Pandas) para contar cuántos datos de Fare superan el límite superior que hemos calculado.



```python
# Contamos cuántos valores de Tarifa superan el límite superior
# df_titanic['fare'] > limite_superior
outliers_fare = df_titanic[columna_analizar > limite_superior]

print(f"Número total de pasajeros con Tarifa atípica: {len(outliers_fare)}")
```

### Conclusiones del Estudio de Atípicos

*   **Detección en `Fare`:** Hemos identificado **[116]** pasajeros con tarifas atípicamente altas. Esto confirma lo que se intuía al revisar el estadístico `.describe()` (donde el máximo era 512, muy por encima del Q3) y el diagrama de caja.
*   **Acción en UD2:** Estos valores atípicos son importantes y deberán ser gestionados en la UD2, ya que podrían sesgar los modelos de regresión si no se manipulan.
