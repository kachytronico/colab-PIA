# Uso de gráficos para mostrar la información relativa a las características del conjunto de datos


```python
sns.boxplot(y=df['flipper_length_mm'])
plt.title('Box Plot of Flipper Length (mm)')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

```python
sns.histplot(y=df['sex'])
plt.title('sex')
plt.ylabel('sex')
plt.show()
```

```python
sns.scatterplot(df, x="bill_length_mm",y="bill_depth_mm")
plt.title('bill_length_mm vs bill_depth_mm')
plt.show()
```

```python
sns.scatterplot(df, x="bill_length_mm",y="bill_depth_mm", hue = df["species"])
plt.title('bill_length_mm vs bill_depth_mm')
plt.show()
```

# Realización de un análisis de correlación entre las características y explicación de su posible correlación.


```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", annot_kws={"size": 10})
plt.show()
```
